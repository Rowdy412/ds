```python
# 6. Turn categorical variables into quantitative variables in Python.
# by using replace() Method
category_mapping = {'Virginica': 0, 'Setosa': 1, 'verseta':2}
df['Species']=df['Species'].replace(category_mapping)
df
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Apply label encoding to the 'species' column
df['Species'] = label_encoder.fit_transform(df['Species'])

# Check updated data types
updated_data_types = df.dtypes
print("Updated Data Types:\n", updated_data_types)
df
```
