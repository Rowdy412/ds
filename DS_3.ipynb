```python
# 2. Write a Python program to display some basic statistical details like percentile, mean, 
# standard deviation etc. of the species of ‘Iris-setosa’, ‘Iris-versicolor’ and ‘Iris-versicolor’ of 
# iris.csv dataset.
# Provide the codes with outputs and explain everything that you do in this step.
```

```python
df = pd.read_csv('Iris1.csv')
df    
```

```python
df=df.fillna(method='ffill')
setosa = df[df.Species=='Setosa']    
Virginica = df[df.Species=='Virginica']
verseta = df[df.Species=='verseta']
```

```python
def statistical_details(species):
    return species.describe()
```

```python
print("Setosa Statistics:")
statistical_details(setosa)
```

```python
print("\nVirginica Statistics:")
statistical_details(Virginica)
```

```python
print("\nverseta Statistics:")
statistical_details(verseta)
```
